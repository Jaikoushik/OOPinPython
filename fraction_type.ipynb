```python
class Fraction:
    def __init__(self, x, y):
        self.num=x
        self.den=y

    def __str__(self):
        return '{}/{}'.format(self.num,self.den)
    
    def __add__(self, other):
        nume = self.num * other.den + self.den * other.num
        deno = self.den * other.den

        return '{}/{}'.format(nume,deno)
    
    def __mul__(self, other):
        nume = self.num*other.num
        deno = self.den*other.den

        return '{}/{}'.format(nume,deno)
    
    def __sub__(self, other):
        nume = self.num * other.den - self.den * other.num
        deno = self.den * other.den

        return '{}/{}'.format(nume,deno)
    
    def __truediv__(self, other):
        nume = self.num*other.den
        deno = self.den*other.num

        return '{}/{}'.format(nume,deno)
    

#So here whenever we try to ""print"" the object the python moves to the __str__ magic method

obj1 = Fraction(3,4)
obj2 = Fraction(4,5)
print(obj1*obj2)
print(obj1+obj2)
print(obj1-obj2)
print(obj1/obj2)
```
