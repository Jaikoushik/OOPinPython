```python
class Hello:

    def __init__(self):
        self.pin=""
        self.balance=0
        self.menu()
        
    def menu(self):
        user = input("""
        1. Create Pin
        2.Change PIn
        3.Check balance""")

        if user == "1":
            self.create_pin()

        elif user == "2":
            self.change_pin()

        elif user == "3":
            self.check_balance()
        elif user == "4":
            self.withdraw_money()

    def create_pin(self):
        user_pin = int(input("Enter the pin"))
        self.pin = user_pin

        if self.pin:
            print("Pin created successfully")
            
            balance = int(input("Enter the balance"))
            self.balance = balance 
            self.menu()
        else:
            print("Not created successfully")
            self.menu()
    
    def change_pin(self):
        old_pin = int(input("Enter old pin"))

        if old_pin == self.pin:
            new_pin = int(input("Enter new pin"))
            self.pin = new_pin
            print("Pin change successful")
            self.menu()
        else:
            print("Not possible")
            self.menu()

    def check_balance(self):
        pin = int(input("Enter your pin"))

        if pin == self.pin:
            print("The balance is ",self.balance)
        else:
            print("Invalid pin")
    
    def withdraw_money(self):
        pin = int(input("Enter the pin"))
        if pin == self.pin:
            withdraw = int(input("Enter the amount"))
            self.balance = self.balance-withdraw
            print("The amount withdrawn successfully the current balance is ",self.balance)
        else:
            print("Invalid Pin")


```
