```python
class Point:

    def __init__(self,x ,y):
        self.x_cod = x
        self.y_cod = y

    def __str__(self):
        return f"<{self.x_cod},{self.y_cod}>"
    
    def euclidean_distance(self,other):
        return ((self.x_cod-other.x_cod)**2+(self.y_cod-other.y_cod)**2)**0.5
    
    def distance_from_origin(self):
        return self.euclidean_distance(Point(0,0))  #self is sent before itself no need to send as a parameter ->self.eucl.....

class Line:

    def __init__(self,A,B,C):
        self.A = A
        self.B = B
        self.C = C

    def __str__(self):
        return '{}x+{}y+{}=0'.format(self.A,self.B,self.C)
    
    def point_on_line(line,point):
        if line.A*point.x_cod + line.B*point.y_cod+line.C == 0:
            return "Lies on the line"
        else:
            return "Doesnt lie on the line"
        
    def shortest_distance(self,other):
        return abs(self.A*other.x_cod + self.B*other.y_cod + self.C)/(self.A**2+self.B**2)




l1 = Line(1,1,-2)
p1 = Point(1,1)
print(l1)
print(p1)

l1.point_on_line(p1)
l1.shortest_distance(p1) 

```
